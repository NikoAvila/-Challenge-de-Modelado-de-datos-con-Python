```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Concatenar los 4 dataframes en uno solo
total_productos = (
    len(tienda) +
    len(tienda2) +
    len(tienda3) +
    len(tienda4)
)

# Cantidad de registros en todas las tablas
print(f"Cantidad total de productos en ambas tablas: {total_productos}")
```

# 2. Ventas por categoría

```python
#uniendo tablas
df_total = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
# Agrupar y sumar las ventas de los productos por categoría
ventas_por_categoria = df_total.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

# Convertir el valor a millones y redondear
ventas_millones = (ventas_por_categoria / 1_000_000).round(2)

# Convertir las tablas para obtener una mejor visualizacion
ventas_df = ventas_millones.reset_index()
ventas_df.columns = ['Categoría del Producto', 'Ventas en Millones ($)']

# Mostrar resultado final
print("Ventas por Categoría (en millones de $):")
print(ventas_df)

```

```python
import matplotlib.pyplot as plt

# Crear gráfico de barras horizontales
plt.barh(ventas_df['Categoría del Producto'], ventas_df['Ventas en Millones ($)'])
plt.xlabel('Ventas en Millones ($)')
plt.title('Ventas por Categoría')
plt.gca().invert_yaxis()  # Ordenar de mayor a menor
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio
calificacion_promedio = df_total['Calificación'].mean()

print(f"Calificación promedio de la tienda: {calificacion_promedio:.2f}")
```

```python
# Formatear para mostrar con 2 decimales para asi mostrar la calificacion promedio por categoria
calificacion_formateada = calificacion_por_categoria.round(2).apply(lambda x: f"{x:.2f}")

# Convertir
calificacion_df = calificacion_formateada.reset_index()
calificacion_df.columns = ['Categoría del Producto', 'Promedio por Calificación']

# Resultado de la calificacion promedio por categoria
print(calificacion_df)
```

# 4. Productos más y menos vendidos

```python
#Productos más y menos vendidos
productos_vendidos = df_total.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)
productos_mas_vendidos = productos_vendidos.head(5)  # Los 5 productos más vendidos
productos_menos_vendidos = productos_vendidos.tail(5)  # Los 5 productos menos vendidos

# dar formato de tabla
productos_mas_vendidos_df = pd.DataFrame(productos_mas_vendidos)
productos_menos_vendidos_df = pd.DataFrame(productos_menos_vendidos)

# Renombrar las columnas para mayor claridad
productos_mas_vendidos_df.columns = ['Cantidad Vendida']
productos_menos_vendidos_df.columns = ['Cantidad Vendida']
```

```python
# Mostrar las tablas de productos más vendidos
print("\nProductos más vendidos:")
print(productos_mas_vendidos_df)
```

```python
# Mostrar las tablas de productos menos vendidos
print("\nProductos menos vendidos:")
print(productos_menos_vendidos_df)
```

# 5. Envío promedio por tienda

```python
# Agrupar por 'Lugar de Compra' (tienda) y calcular el promedio de 'Costo de envío'
envio_promedio_tienda = df_total.groupby('Lugar de Compra')['Costo de envío'].mean()

# Redondear el costo del envio
envio_promedio_tienda_redondeado = envio_promedio_tienda.round(0)

# Resultado en formato de una tabla
envio_promedio_tienda_df = pd.DataFrame(envio_promedio_tienda_redondeado)

# Renombrar las columnas para indicar el envio promedio
envio_promedio_tienda_df.columns = ['Costo Envío Promedio $']

# Mostrar la tabla
print("Envío promedio por tienda:")
print(envio_promedio_tienda_df)

```
